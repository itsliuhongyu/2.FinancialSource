For this operation, I will be creating a giant csv file combining all the information from 2015-2022

```python
import pandas as pd

df1 = pd.read_csv('./FECdata/2015-2016.csv')
df2 = pd.read_csv('./FECdata/2017-2018.csv')
df3 = pd.read_csv('./FECdata/2019-2020.csv')
df4 = pd.read_csv('./FECdata/2021-2022.csv')
```

```python
concatenated_df = pd.concat([df1, df2], axis=0)
concatenated_df
```

```python
concatenated_df = pd.concat([concatenated_df, df3], axis=0)
concatenated_df = pd.concat([concatenated_df, df4], axis=0)
concatenated_df
```

```python
concatenated_df.to_csv('fulldata.csv', index=False)
```
